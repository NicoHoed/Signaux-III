```python
# Import dependencies
import os, sys
import matplotlib.pyplot as plt
from skimage import io

# Add parent directory to the Python path so 'src' can be found
sys.path.append(os.path.abspath('..'))

# Import the preprocessing function from preprocessing.py
from src.preprocessing import preprocess_image

# Load the input image and apply preprocessing to obtain a binary image
img_path = '../data/img/INT-qwerty-1.jpg'
binary = preprocess_image(io.imread(img_path))

# Display the resulting binary image in grayscale
plt.imshow(binary, cmap='gray')
plt.title('Binarized Image after Preprocessing')
plt.axis('off')
plt.show()
```
